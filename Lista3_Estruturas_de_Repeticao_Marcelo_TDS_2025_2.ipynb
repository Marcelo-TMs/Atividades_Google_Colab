```python
'''
9. Elabore um algoritmo que solicite ao usuário um número inteiro que indicará a quantidade
de vezes que o texto "Hello World!" será impresso na tela, um em cada linha.
'''

num = int(input("Quantas vezes (número INTEIRO) você deseja que a frase 'Hello World!' apareça na tela? "))

for i in range(num):
  print("Hello World!")
  i -= 1
```

```python
'''
10. Elabore um algoritmo que solicite ao usuário uma palavra e um número inteiro que
indicará a quantidade de vezes que a palavra digitada será impressa na tela, um em cada
linha.
'''

palavra = str(input("Digite uma palavra: "))
num = int(input(f"Quantas vezes (número INTEIRO) você deseja que a frase {palavra} apareça na tela? "))

for i in range(num):
  print(palavra)
  i -= 1
```

```python
'''
11. Elabore um algoritmo que leia um número de entrada que indicará a quantidade de
números a serem lidos. Em seguida, leia n números (conforme o valor informado
anteriormente) e imprima a soma e a média aritmética dos números informados.
'''

entrada = int(input("Quantos números você deseja inserir? "))
soma = 0

for i in range(entrada):
  numero = int(input("Digite um número: "))
  soma += numero
  i += 1

media = soma / entrada

print("\nA soma dos números é:", soma)
print("A média dos números é:", media)
```

```python
'''
12. Elabore um algoritmo que leia um número de entrada que indicará a quantidade de
registros a serem lidos (N). Em seguida algoritmo deve solicitar o nome e idade de N
pessoas e ao final apresentar o nome da pessoa mais velha.
'''

entrada = int(input("Quantos registros você deseja inserir? "))
num = 1
maior = 0

for i in range(entrada):
  nome = input(f"\nDigite o nome da {num}° pessoa: ")
  idade = int(input("Digite a idade dessa pessoa: "))
  num += 1

  if idade > maior:
    maior = idade
    pessoa = nome

print("\nA pessoa mais velha é:", pessoa)
```

```python
'''
13. Elabore um algoritmo que leia um número de entrada que indicará a quantidade de
registros a serem lidos (N). Em seguida algoritmo deve solicitar o sexo (M/F) e idade de N
pessoas e ao final apresentar a média de idade de ambos os gêneros catalogados.
'''

entrada = int(input("Quantos registros você deseja inserir? "))
num = 0
masculino = 0
feminino = 0
media_M = 0
media_F = 0

while num < entrada:
  sexo = input("\nSexo da pessoa (M/F): ")

  if sexo == "M":
    masculino += 1

    idade = int(input("Idade dessa pessoa: "))
    media_M += idade
  elif sexo == "F":
    feminino += 1

    idade = int(input("Idade dessa pessoa: "))
    media_F += idade
  else:
    print("Informação inserida é inválida!")

  num += 1

if masculino == 0:
  media_M = 0
else:
  media_M = media_M / masculino

if feminino == 0:
  media_F = 0
else:
  media_F = media_F / feminino

print("\nA média de idade masculina é:", media_M)
print("A média de idade feminina é:", media_F)
```

```python
'''
14. Elabore um algoritmo que solicite ao usuário 10 números reais e ao final apresente o
maior e o menor deles.
'''

maior = 0
menor = 0

for i in range(1, 11):
  num = float(input(f"Digite o {i}° número: "))

  if i == 1:
    maior = num
    menor = num
  else:
    if num > maior:
      maior = num
    if num < menor:
      menor = num

print("O maior número é:", maior)
print("O menor número é:", menor)
```

```python
'''
15. Elabore um algoritmo que solicite N números reais e quando o usuário informar o valor
nulo 0 (zero) o programa ordene e mostre todos os números informados de forma crescente.
'''

print("Obs: Caso digite o número 0, o programa será fechado.")
num = 11
numeros = []

while num != 0:
  num = float(input("Digite um número REAL: "))

  if num == 0:
    break
  else:
    numeros.append(num)

numeros.sort()

for num in numeros:
  print(num)
```

```python
'''
16. Escreva um programa que vá solicitando as idades dos alunos da sala até que todos
sejam informados (perguntar ao usuário se deseja informar a idade do próximo aluno). Ao
final apresentar a idade do mais novo, a idade do mais velho, Quantos alunos têm mais de
18 anos, quantos alunos têm até 18 anos, a média aritmética e a mediana.
'''

entrada = "S"
maior = 0
menor = 0
alunos = 0
maior_18 = 0
menor_18 = 0

while entrada != "N":
  idade = int(input("Digite a idade do aluno: "))
  if idade <= 0:
    print("Informação inserida é inválida!\n")
  else:
    alunos += 1

    if idade > 18:
      maior_18 += 1
    else:
      menor_18 += 1

    if alunos == 1:
      maior = idade
      menor = idade
    else:
      if idade > maior:
        maior = idade
      if idade < menor:
        menor = idade

    entrada = str(input("\nDeseja digitar a idade do próximo aluno? (S/N) "))

media = (maior + menor) / alunos
mediana = (maior + menor) / 2

print("\nIdade do mais novo:", menor)
print("Idade do mais velho:", maior)
print("Quantidade de alunos com mais de 18 anos:", maior_18)
print("Quantidade de alunos com até 18 anos:", menor_18)
print("Média aritmética:", media)
print("Mediana:", mediana)
```
